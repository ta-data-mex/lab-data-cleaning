## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('weather-raw.csv', low_memory=False)
data
```

```python
col_null = data.isnull().sum()
col_null
```

```python
col_alv = list(col_null[col_null > 17].index)
col_alv
```

```python
dataset = data.drop(col_alv, axis=1)
dataset
```

```python
col_null1 = dataset[(dataset['d5'].isnull() == True)]
print(col_null1)
```

```python
col_null2 = dataset[(dataset['d27'].isnull() == True)]
col_null2
```
