## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
messy = pd.read_csv('weather-raw.csv')
```

```python
#Let's take a look at the df:
messy
```

```python
messy.shape
```

```python
ds = []
for i in range(1,32):
    ds.append('d'+str(i))
massy = pd.melt(messy, id_vars=['element', 'month', 'year', 'id']) 
#Melt the df so we get it organize for month instad of day values.
```

```python
data_weather_mx = massy[massy['value'].notnull()] #Filter every row so days without values are erased
data_weather_mx = data_weather_mx.sort_values(by=['month', 'variable']) #sort the dataset
data_weather_mx.columns = ['var', 'month', 'year', 'index', 'Day', 'value'] #change value order (Missed here)
data_weather_mx = data_weather_mx[['index','year','month','Day', 'var', 'value']] #change value order right
data_weather_mx['Day'] = data_weather_mx['Day'].apply(lambda x: x.replace('d', '')) #erase 'd' from day date 
data_weather_mx['Day'] = data_weather_mx['Day'].apply(lambda x: int(x)) #convert it to string so it can be sorted
data_weather_mx = data_weather_mx.sort_values(by=['month', 'Day']) #well sorted
data_weather_mx = data_weather_mx.reset_index() #reset the index level so it gets to the new order
data_weather_mx = data_weather_mx.drop(columns='level_0') #drop the previous index level
data_weather_mx #DF sorted by month, day and with both values available.
data_weather_mx = data_weather_mx.pivot_table( #Pivot the DF on 'var' values while keeping sorted columns right
    values='value',
    index=['index', 'year', 'month', 'Day'],
    columns='var'
    )
#This presentation has similar values grouped, which I like:
data_weather_mx = data_weather_mx.sort_values(by=['month', 'Day'])
data_weather_mx
```
